```python
import numpy as np
import pandas as pd

data = pd.read_excel("2024_01_26.xlsx", sheet_name = '複雜度')
data.head(32)
```

```python
x_row = np.arange(5, 55, 5)
x_row
```

## 1. Compare between different order size

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5), dpi=300)
plt.plot(x_row, data.iloc[28, 1:],marker='o', linestyle='--', label='Variables') # Number of Variables
plt.plot(x_row, data.iloc[29, 1:], marker='s', linestyle='--', label='Constraints') # Number of Constraints
plt.xticks(x_row)

plt.title('Constraints and Variables Counts Across Different Order Size', size=15)
plt.xlabel('Number of Order', size=10)
plt.ylabel('Number of Constraints and Variables', size=10)
plt.legend(loc='lower right')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5), dpi=300)
plt.plot(x_row, data.iloc[31, 1:],marker='o', linestyle='--') # Runtime
plt.xticks(x_row)
plt.ylim(0, 2200)

plt.title('Runtime Across Different Order Size', size=15)
plt.xlabel('Number of Order', size=10)
plt.ylabel('Runtime (sec)', size=10)

plt.show()
```

## 2. Compare between different model size

```python
M = ['(10, 5, 7)', '(30, 15, 20)', '(50, 25, 33)']
V = np.array([3898, 80730, 353642])
C = np.array([3652, 93957, 432942])

plt.figure(figsize=(10, 5), dpi=300)
plt.plot(M, V, label='Variables') # Number of Variables
plt.plot(M, C, label='Constraints') # Number of Constraints
plt.xticks(M)

plt.title('Constraints and Variable Counts Across Different Model Scales', size=15)
plt.xlabel('Model', size=10)
plt.ylabel('Count', size=10)
plt.legend(loc='lower right')

plt.show()
```

```python
M = ['(10, 5, 7)', '(30, 15, 20)', '(50, 25, 33)']
R = np.array([337, 388, 1177])

plt.figure(figsize=(10, 5), dpi=300)
plt.plot(M, R) # Runtime
plt.xticks(M)

plt.title('Runtime Across Different Model Scales', size=15)
plt.xlabel('Model', size=10)
plt.ylabel('Runtime (sec)', size=10)

plt.show()
```
