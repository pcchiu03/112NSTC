# 穩定度測試

```python
import numpy as np
import pandas as pd

data = pd.read_excel("2024_02_22.xlsx", sheet_name = '穩定度')
data.head(32)
```

```python
x_row = np.arange(1, 11, 1)

x_row
```

```python
avg = sum(data.iloc[31, 1: 11]) / len(data.iloc[31, 1: 11])

avg
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4), dpi=300)
plt.plot(x_row, data.iloc[31, 1: 11], marker='o', linestyle='None')
plt.axhline(avg, color='r', linestyle='--', label='Average Runtime')
plt.xticks(x_row)
plt.ylim(0, 349)

plt.title('Runtime Across Different Dataset', size=15)
plt.xlabel('Dataset', size=10)
plt.ylabel('Runtime (sec)', size=10)
plt.legend()

plt.show()
```
